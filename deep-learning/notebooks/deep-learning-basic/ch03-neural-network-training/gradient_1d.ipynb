```python
# coding: utf-8
import numpy as np
import matplotlib.pylab as plt
```

```python
# 수치 미분 함수
def numerical_diff(f, x):
    h = 1e-4 #0.0001
    return ( f(x+h) - f(x-h) ) / (2*h)
```

```python
# 미분할 함수
def function_1(x):
    return 0.01 * x * 2 + 0.1 * x
```

```python
# 접선 구하는 함수
def tangent_line(f, x):
    d = numerical_diff(f,x) #수치미분으로 기울기 구하기
    print(d)
    y = f(x) - d*x 
    return lambda t: d*t + y
```

```python
x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)
plt.xlabel("x")
plt.ylabel("f(x)")

tf = tangent_line(function_1, 5)
y2 = tf(x)

plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```
