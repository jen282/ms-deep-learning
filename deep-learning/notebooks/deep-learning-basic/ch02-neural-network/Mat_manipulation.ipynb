```python
import numpy as np
```

1. 행렬 살펴보기

* ndim(행렬) : 차원수 확인
* shape : 형상

```python
A = np.array([1,2,3,4])
print('행렬: ', A)
print('차원: ', np.ndim(A))
print('행/열: ', A.shape)
```

```python
B = np.array([[1,2], [3,4], [5,6]])
print('행렬:\n',B)
print('차원: ', np.ndim(B))
print('행/열: ', B.shape)
```

2. 행렬 계산하기

```python
# 곱셈1
A = np.array([[1,2],[3,4]])
print('행/열(A): ', A.shape)

B = np.array([[5,6],[7,8]])
print('행/열(B): ', B.shape)

# 행렬 곱 연산
np.dot(A, B)
```

```python
# 곱셈2
A = np.array([[1,2,3],[4,5,6]])
print('행/열(A): ', A.shape)
print('A:')
print(A)

B = np.array([[1,2], [3,4], [5,6]])
print('행/열(B): ', B.shape)
print('B:')
print(B)

# 행렬 곱 연산
np.dot(A, B)
```

3. 신경망에서의 행렬의 곱 
신경망의 x값들을 첫번째 행렬, 가중치를 두번째 행렬로 가정하고 두 행렬의 곱을 계산한 결과는 그 결과인 y값들과 같다


```python
# 신경망에서의 행렬의 곱 
X = np.array([1,2])
print('X:')
print(X)

W = np.array([[1, 3, 5], [2, 4, 6]])
print('W:')
print(W)

Y = np.dot(X, W)
print('결과:')
print(Y)
```
